***Second Highest Salary***


```python
Employee = pd.DataFrame([], columns=['id', 'salary']).astype({'id':'int64', 'salary':'int64'})
```

```python
id = [1]
salary = [100]
Employee['id'] = id
Employee['salary'] = salary
```

```python
Employee = Employee.drop_duplicates(subset = 'salary', keep = 'first', inplace = False)
Employee
```

```python
sorted_salary = Employee['salary'].sort_values(ascending = False)
sorted_salary
```

```python
if len(sorted_salary) > 1:
  sec_max = sorted_salary[1]
else:
  sec_max = None
```

```python
res = pd.DataFrame({'SecondHighestSalary': [sec_max]})
res
```

```python
n = 1
sorted_salary = Employee['salary'].sort_values(ascending = False)
sorted_salary = list(sorted_salary.drop_duplicates(inplace = False))
sorted_salary
```

***Rank Scores***

```python
Scores = pd.DataFrame([], columns=['id', 'score']).astype({'id':'int64', 'score':'float64'})
id = [1,2,3,4,5,6]
scores = [3.5,3.65,4.0,3.85,4,3.65]
Scores['id'] = id
Scores['score'] = scores
Scores
```

```python
Scores['rank'] = Scores['score'].rank(method = 'dense' , ascending = False).astype('int')
Scores = Scores[['score' , 'rank']].sort_values('score' , ascending = False)
Scores
```

***Department Highest Salary***

```python
employee = pd.DataFrame([], columns=['id', 'name', 'salary', 'departmentId']).astype({'id':'int64', 'name':'object', 'salary':'int64', 'departmentId':'int64'})
department = pd.DataFrame([], columns=['id', 'name']).astype({'id':'int64', 'name':'object'})
```

```python
id = [1,2,3,4,5]
name = ['joe','jim','henry','sam','max']
salary = [70000,90000,80000,60000,90000]
departmentId = [1,1,2,2,1]
employee['id'] = id
employee['name'] = name
employee['salary'] = salary
employee['departmentId'] = departmentId
employee
```

```python
id = [1,2]
name = ['IT','Sales']
department['id'] = id
department['name'] = name
department
```

```python
data = pd.merge(employee, department, left_on = 'departmentId', right_on = 'id', how = 'left')
data
```

```python
df = data.groupby('name_y').apply(lambda x: x[x.salary == x.salary.max()])
print(df)
```

```python
df = df.rename(columns={'name_y': 'Department', 'name_x': 'Employee'})

# Drop the extra 'id' columns and keep only the necessary columns
df = df.drop(columns=['id_x', 'id_y'])
```

```python
df = df.reset_index()
df
```

```python
df = df[['Department','Employee','salary']]
df
```

**Big Countries**

```python
world = pd.DataFrame([], columns=['name', 'continent', 'area', 'population', 'gdp']).astype({'name':'object', 'continent':'object', 'area':'int64', 'population':'int64', 'gdp':'int64'})
name = ['afg','alb','alg','and','ang']
continent = ['asia','eu','afr','eu','afr']
area = [65000,78000,450,345,678]
population = [25550000,4567890,9887,677668,9887]
gdp = [1,2,3,4,5]
world['name'] = name
world['continent'] = continent
world['area'] = area
world['population'] = population
world['gdp'] =  gdp
world
```

***Classes have more than 5 students***

```python
courses = pd.DataFrame([], columns=['student', 'class']).astype({'student':'object', 'class':'object'})
stud = ['A','B','C','D','E','F','G','H','I']
clas = ['Math','English','Math','Biology','Math','Computer','Math','Math','Math']
courses['student'] = stud
courses['class'] = clas
courses
```

```python
df = courses.groupby('class')['student'].count().reset_index()
df = df[df['student'] >= 5]
df = df[['class']]
df
```

***Actors and Directors who cooperated with atleast three times***

```python
actor_dir = pd.DataFrame([], columns=['actor_id', 'director_id', 'timestamp']).astype({'actor_id':'int64', 'director_id':'int64', 'timestamp':'int64'})
actor = [1,1,1,1,1,2,2]
director = [1,1,1,2,2,1,1]
ts = [0,1,2,3,4,5,6]
actor_dir['actor_id'] = actor
actor_dir['director_id'] = director
actor_dir['timestamp'] = ts
actor_dir
```

```python
df = actor_dir.groupby(['actor_id','director_id']).agg(count =('director_id','count')).reset_index()
df = df[df['count'] >= 3]
df = df[['actor_id','director_id']]
df
```

***Game Play Analysis***

```python
data = {
    'player_id': [1, 1, 2, 3, 3],
    'device_id': [2, 2, 3, 1, 4],
    'event_date': ['2016-03-01', '2016-05-02', '2017-06-25', '2016-03-02', '2018-07-03'],
    'games_played': [5, 6, 1, 0, 5]
}

activity = pd.DataFrame(data)

# Optionally, convert the 'event_date' column to a datetime type
activity['event_date'] = pd.to_datetime(activity['event_date'])

activity
```

```python
df = activity.groupby('player_id').agg(first_login = ('event_date','min')).reset_index()
df
```

***Article Views 1***

```python
data = {
    'article_id': [1, 1, 2, 2, 4, 3, 3],
    'author_id': [3, 3, 7, 7, 7, 4, 4],
    'viewer_id': [5, 6, 7, 6, 1, 4, 4],
    'view_date': ['2019-08-01', '2019-08-02', '2019-08-01', '2019-08-02', '2019-07-22', '2019-07-21', '2019-07-21']
}

# Create the pandas DataFrame
views = pd.DataFrame(data)

# Convert the 'view_date' column to pandas datetime format
views['view_date'] = pd.to_datetime(views['view_date'])
views
```

```python
views = views[views['author_id'] == views['viewer_id']]
views = views[['author_id']].sort_values(by = 'author_id')
views
```

```python
views.drop_duplicates(subset = 'author_id', inplace = True)
views.rename(columns = {'author_id':'id'}, inplace = True)
```
