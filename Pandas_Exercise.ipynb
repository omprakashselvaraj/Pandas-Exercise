***Group Sold Products by the date***

```python
activities = pd.DataFrame([], columns=['sell_date', 'product']).astype({'sell_date':'datetime64[ns]', 'product':'object'})
selldate = ['2020-05-30', '2020-06-01','2020-06-02','2020-05-30','2020-06-01','2020-06-02','2020-05-30']
products = ['Headphone','Pencil','Mask', 'Basketball','Bible','Mask', 'T-shirt']
activities['sell_date'] = selldate
activities['product'] = products
activities
```

```python
def distinct_word(x):
  res = ''
  x = list(set(x.split(',')))
  x.sort()
  for i in range(len(x)):
    if i == len(x)-1:
      res = res + x[i]
    else:
      res = res + x[i] + ','
  return res
```

```python
def count_fun(x):
  x = x.split(',')
  return len(x)
```

```python
df1 = activities.groupby('sell_date')['product'].apply(lambda x: ','.join(x)).reset_index()
df1['product'] = df1['product'].apply(distinct_word)
df1['num_sold'] = df1['product'].apply(count_fun)
df1
```

```python
df1 = df1[['sell_date','num_sold','product']]
df1
```

***Patients with a condition***

```python
data = {
    'patient_id': [1, 2, 3, 4, 5],
    'patient_name': ['Daniel', 'Alice', 'Bob', 'George', 'Alain'],
    'conditions': ['YFEV COUGH', '', 'DIAB100 MYOP', 'ACNE DIAB100', 'DIAB201']
}

# Create the pandas DataFrame
patients = pd.DataFrame(data)
patients
```

```python
def find_diab(x):
  x = x.split(" ")
  for i in x:
    if i.startswith("DIAB1"):
      return 1
  return 0
```

```python
patients['val'] = patients['conditions'].apply(find_diab)
patients
```

```python
patients = patients[patients['val'] == 1]
patients = patients[['patient_id','patient_name','conditions']]
patients
```

***Find total time spent by each employee***

```python
data = {
    'emp_id': [1, 1, 1, 2, 2],
    'event_day': ['2020-11-28', '2020-11-28', '2020-12-03', '2020-11-28', '2020-12-09'],
    'in_time': [4, 55, 1, 3, 47],
    'out_time': [32, 200, 42, 33, 74]
}

# Create a pandas DataFrame
employees = pd.DataFrame(data)
employees['event_day'] = pd.to_datetime(employees['event_day'])
# Display the DataFrame
print(employees)
```

```python
employees['total_time'] = employees['out_time'] - employees['in_time']
employees
```

```python
df = employees.groupby(['emp_id','event_day']).agg(total_time = ('total_time','sum')).reset_index()
df['day'] = df['event_day'].dt.strftime('%Y-%m-%d')
df = df[['day','emp_id','total_time']]
df = df.sort_values(['emp_id','day'])
df
```

***Calculate special bonus***

```python
data = {
    'employee_id': [2, 3, 7, 8, 9],
    'name': ['Meir', 'Michael', 'Addilyn', 'Juan', 'Kannon'],
    'salary': [3000, 3800, 7400, 6100, 7700]
}

# Create a pandas DataFrame
employees = pd.DataFrame(data)
employees
```

```python
employees['bonus'] = employees[(~employees['name'].str.startswith('M')) & (employees['employee_id'] % 2 == 1)]['salary']
employees['bonus'] = employees.bonus.fillna(value=0)
employees = employees[['employee_id', 'bonus']].sort_values(by='employee_id')
employees
```
